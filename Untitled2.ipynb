1. Загрузите файл classification.csv. В нем записаны истинные классы объектов выборки (колонка true) и ответы некоторого классификатора (колонка predicted).

```python
import pandas as pd
from sklearn.metrics import roc_auc_score, precision_recall_curve

data = pd.read_csv('/content/classification.csv')
tru = data.iloc[:, 0].values
pred = data.iloc[:, 1].values
```

2. Заполните таблицу ошибок классификации. Для этого подсчитайте величины TP, FP, FN и TN согласно их определениям. Ответ в данном вопросе — четыре числа через пробел.

```python
TP = ((tru == 1) & (pred == 1)).sum()
FP = ((tru == 0) & (pred == 1)).sum()
FN = ((tru == 1) & (pred == 0)).sum()
TN = ((tru == 0) & (pred == 0)).sum()

print(f"TP = {TP}")
print(f"FP = {FP}")
print(f"FN = {FN}")
print(f"TN = {TN}")
print(f"\n{TP} {FP} {FN} {TN}")
```

3. Посчитайте основные метрики качества классификатора: Accuracy, Precision, Recall, F-мера

```python
acc = (TP + TN) / (TP + TN + FP + FN)
print("Accuracy:", round(acc, 3))

prec = TP / (TP + FP) if (TP + FP) > 0 else 0
print("Precision:", round(prec, 3))

rec = TP / (TP + FN) if (TP + FN) > 0 else 0
print("Recall:", round(rec, 3))

f1 = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0
print("F1-score:", round(f1, 3))
```

4. Имеется четыре обученных классификатора. В файле scores.csv записаны истинные классы и значения степени принадлежности положительному классу для каждого классификатора на некоторой выборке: • для логистической регрессии — вероятность положительного класса (колонка score_logreg), • для SVM — отступ от разделяющей поверхности (колонка score_svm), • для метрического алгоритма — взвешенная сумма классов соседей (колонка score_knn), • для решающего дерева — доля положительных объектов в листе (колонка score_tree)

```python
data2 = pd.read_csv('scores.csv')
tru_scores = data2['true']

models = ['score_logreg', 'score_svm', 'score_knn', 'score_tree']
auc_scores = {}

for model in models:
    y_score = data2[model]
    auc = roc_auc_score(tru_scores, y_score)
    auc_scores[model] = auc
    print(f"AUC-ROC для {model:15s} = {auc:.4f}")
```

```python
best_auc_model = max(auc_scores, key=auc_scores.get)
print(f"\nНаибольшее значение AUC-ROC у классификатора: {best_auc_model}")
```

6. Какой классификатор достигает наибольшей точности (Precision) при полноте (Recall) не менее 70% (укажите название столбца с ответами этого классификатора)? Какое значение точности при этом получается?

```python
best_prec = -1
best_prec_name = ''
for name in models:
    prec_curve, rec_curve, _ = precision_recall_curve(tru_scores, data2[name])
    # Ищем максимальную precision среди точек с recall >= 0.7
    mask = rec_curve >= 0.7
    if mask.any():
        max_prec = max(prec_curve[mask])
        if max_prec > best_prec:
            best_prec = max_prec
            best_prec_name = name

print(f"Лучший по Precision (при Recall>=0.7): {best_prec_name}")
print(f"Precision = {round(best_prec, 2)}")
```

2. На загруженных данных обучите преобразование PCA с числом компоненты равным 10. Скольких компонент хватит, чтобы объяс- нить 90% дисперсии?

```python
pca = PCA(n_components=10)
pca.fit(data_stocks)

expl_var = pca.explained_variance_ratio_
print("Объяснённая дисперсия по компонентам (%):")
for i, val in enumerate(expl_var, 1):
    print(f"Компонента {i}: {val*100:.2f}%")

cumsum = np.cumsum(expl_var)
n_components_90 = np.argmax(cumsum >= 0.90) + 1

print("Накопленная доля дисперсии по компонентам:")
for i, val in enumerate(cumsum, 1):
    print(f"  {i} компонент(а): {val*100:.2f}%")
print(f"\nДля объяснения 90% дисперсии достаточно {n_components_90} компонент(ы)")
```

1. Загрузите данные из файла abalone.csv. Это датасет, в котором тре- буется предсказать возраст ракушки (число колец) по физическим измерениям.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import r2_score

data = pd.read_csv('abalone.csv')
```

2. Преобразуйте признак Sex в числовой: значение F должно перейти в -1, I — в 0, M — в 1.

```python
sex_mapping = {'F': -1, 'I': 0, 'M': 1}
data['Sex'] = data['Sex'].map(sex_mapping)
```

3. Разделите содержимое файлов на признаки и целевую переменную. В последнем столбце записана целевая переменная, в остальных — признаки.

```python
x = data.drop('Rings', axis=1).values
y = data['Rings'].values
```

4. Обучите случайный лес (sklearn.ensemble.RandomForestRegressor)
с различным числом деревьев: от 1 до 50 (random_state=1). Для каждого из вариантов оцените качество работы полученного леса
на кросс-валидации по 5 блокам. Используйте параметры "random_state=1"и "shuffle=True"при создании генератора кросс- валидации sklearn.cross_validation.KFold. В качестве меры качества воспользуйтесь коэффициентом детерминации (sklearn.metrics.r2_score).

```python
n_folds = 5
kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)

n_trees_range = range(1, 51)
mean_r2_scores = []

for n in n_trees_range:
    rf = RandomForestRegressor(n_estimators=n, random_state=1, n_jobs=-1)
    scores = cross_val_score(rf, x, y, cv=kf, scoring='r2')
    mean_r2 = np.mean(scores)
    mean_r2_scores.append(mean_r2)

    if n % 10 == 0:
        print(f"n_estimators={n}, средний R² = {mean_r2:.4f}")
```

5. Определите, при каком минимальном количестве деревьев случай- ный лес показывает качество на кросс-валидации выше 0.52. Это количество и будет ответом на задание.

```python
min_trees = None
for n, r2 in zip(n_trees_range, mean_r2_scores):
    if r2 > 0.52:
        min_trees = n
        break

print(f"Минимальное количество деревьев: {min_trees}")
```

6. Обратите внимание на изменение качества по мере роста числа деревьев. Ухудшается ли оно?


```python
if all(mean_r2_scores[i] <= mean_r2_scores[i+1] + 0.005 for i in range(len(mean_r2_scores)-1)):
    print("Качество не ухудшается – оно растёт или стабилизируется.")
else:
    print("Наблюдаются колебания, но в целом тренд – улучшение или стабилизация.")
```
